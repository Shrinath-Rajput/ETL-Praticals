```python
#Import sql

import sqlite3
```

```python
#Connect database

database=sqlite3.connect('Pratical.db')
print(database)
cursor=database.cursor()
```

```python
#Create Table 

cursor.execute("""
CREATE TABLE IF NOT EXISTS employees(
    id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    age INTEGER,
    department TEXT
)
""")
```

```python
#commit 

database.commit()
```

```python
# insert value in table 

cursor.execute("INSERT INTO employees (id,name,age,department) VALUES (1,'Rahul',25,'IT')")
cursor.execute("INSERT INTO employees (id,name,age,department) VALUES (2,'Amit',28,'HR')")
cursor.execute("INSERT INTO employees (id,name,age,department) VALUES (3,'Sneha',24,'Finance')")
cursor.execute("INSERT INTO employees (id,name,age,department) VALUES (4,'Rohit',30,'IT')")
cursor.execute("INSERT INTO employees (id,name,age,department) VALUES (5,'Priya',26,'Marketing')")
```

```python
#display table 

cursor.execute("SELECT * FROM employees")
rows = cursor.fetchall()

for row in rows:
    print(row)
```

```python
#Update value in table 

cursor.execute("UPDATE employees SET department='Admin' WHERE id=1")
```
