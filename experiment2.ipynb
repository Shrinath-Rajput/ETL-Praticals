```python
#i. Find all hyperlinks present on the webpage

import requests
from bs4 import BeautifulSoup

url = "https://www.w3.org/WAI/ER/tests/xhtml/testfiles/resources/pdf/dummy.pdf"
response = requests.get(url)

soup = BeautifulSoup(response.text, "html.parser")

links = soup.find_all("a")

print("All hyperlinks from W3Schools:\n")

for link in links:
    print(link.get("href"))
```

```python
#ii. Check whether PDF files exist on the webpage

import requests
from bs4 import BeautifulSoup

url = "https://www.w3.org/WAI/ER/tests/xhtml/testfiles/resources/pdf/dummy.pdf"
response = requests.get(url)

soup = BeautifulSoup(response.text, "html.parser")

print("PDF links found:\n")

for link in soup.find_all("a"):
    href = link.get("href")
    if href and ".pdf" in href:
        print(href)
```

```python
##iii. Extract and download the available PDF file(s)

import requests
import os

pdf_url = "https://www.w3.org/WAI/ER/tests/xhtml/testfiles/resources/pdf/dummy.pdf"

os.makedirs("pdf_files", exist_ok=True)

pdf_name = "sample.pdf"
response = requests.get(pdf_url)

with open("pdf_files/" + pdf_name, "wb") as f:
    f.write(response.content)

print("PDF downloaded successfully.")
```
